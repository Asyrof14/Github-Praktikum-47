```python
from collections import deque

# 1. Buat graf tak berarah dengan minimal 5 simpul dan 7 sisi
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C', 'E'],
    'C': ['A', 'B', 'D'],
    'D': ['A', 'C', 'E'],
    'E': ['B', 'D']
}


def print_graph(g):
    for node in g:
        print(f"{node}: {g[node]}")

# 2. Implementasi BFS dan DFS
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=' ')
            visited.add(vertex)
            queue.extend([v for v in graph[vertex] if v not in visited])
    print()

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=' ')
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)


def bfs_visit_order(graph, start):
    visited = set()
    queue = deque([start])
    order = []
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            order.append(vertex)
            queue.extend([v for v in graph[vertex] if v not in visited])
    return order


def find_path(graph, start, end, path=None):
    if path is None:
        path = []
    path = path + [start]
    if start == end:
        return path
    for node in graph[start]:
        if node not in path:
            new_path = find_path(graph, node, end, path)
            if new_path:
                return new_path
    return None


def is_connected(graph):
    start_node = next(iter(graph))  # Ambil salah satu node
    visited = set()

    def dfs_inner(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs_inner(neighbor)

    dfs_inner(start_node)
    return len(visited) == len(graph)


print("1. Print Graph:")
print_graph(graph)

print("\n2. BFS dari A:")
bfs(graph, 'A')

print("DFS dari A:")
dfs(graph, 'A')
print()

print("\n3. BFS Visit Order dari A:")
print(bfs_visit_order(graph, 'A'))

print("\n4. Path dari A ke E:")
print(find_path(graph, 'A', 'E'))

print("\n5. Is graph connected?")
print(is_connected(graph))
```
