```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

```python
from pyspark.sql import SparkSession
import pandas as pd

# Pyspark
spark = SparkSession.builder.appName("BigDataPractice_5").getOrCreate()

data_pyspark = [("Asyrof", 19, "Game Director", "Bermain Game", "L"), ("Budi", 23, "Pegawai Kantoran", "Berkebun", "L"), ("Citra", 25, "Programmer", "Ngoding", "P"), ("Dhimas", 30,"Guru" , "Silat", "L")]
columns_pyspark = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df_pyspark = spark.createDataFrame(data_pyspark, columns_pyspark)

# KOnversi
df_pyspark_to_pandas = df_pyspark.toPandas()

# Pandas
new_data_pandas = {"Nama": ["Asyrof", "Budi"], "Kota": ["Magelang", "Yogyakarta"]}
df_new_pandas = pd.DataFrame(new_data_pandas)

df_merged = pd.merge(df_pyspark_to_pandas, df_new_pandas, on="Nama")
print("DataFrame yang digabungkan dari PySpark dan Pandas:")
print(df_merged)

rata_rata_usia = df_merged['Usia'].mean()
print(f"\nRata-rata usia dari data yang digabungkan: {rata_rata_usia}")
```

```python
from pyspark.sql import SparkSession
import pandas as pd

# PySpark
spark = SparkSession.builder.appName("BigDataPractice_6").getOrCreate()

data_pyspark = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns_pyspark = ["Nama", "Usia"]
df_pyspark = spark.createDataFrame(data_pyspark, columns_pyspark)

# Konversi
df_pyspark_to_pandas = df_pyspark.toPandas()

# Pandas
data_pandas= {"Nama": ["Eko", "Fani"], "Usia": [50, 25]}
df_pandas = pd.DataFrame(data_pandas)

# Menggabungkan DataFrame
df_combined = pd.concat([df_pyspark_to_pandas, df_pandas], ignore_index=True)
print("DataFrame setelah penggabungan:")
print(df_combined)

# Nilai maksimum usia
max_usia = df_combined['Usia'].max()
print(f"\nNilai maksimum usia dari data yang digabungkan: {max_usia}")
```
